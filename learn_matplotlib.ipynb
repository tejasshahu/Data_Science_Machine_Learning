## Matplotlib Example Workflow

We want:
* Line Plot
* Bar Plot
* Scatter Plot
* Histogram
* Subplots

```python
# 0. import matplotlib and get it ready to plot in your jupyter notebook
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare Data
x = [1, 2, 3, 4]
y= [11, 22, 33, 44]

# 2. Setup Plot
fig, ax = plt.subplots(figsize=(10, 10)) # width, height

# 3. Plot Data
ax.plot(x, y)

# 4. Customize Plot
ax.set(title="Simple Plot",
      xlabel="X-Axis",
      ylabel="Y-Asix")

# 5. Save and Show
fig.savefig("images/sample-plot.png")
```

## Two options for subplots

```python
# subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, 
                                             ncols=2, 
                                             figsize=(10, 5))
# plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.randn(10), np.random.randn(10));
ax3.bar(list(nut_butter_sales.keys()), list(nut_butter_sales.values()));
ax4.hist(np.random.randn(1000));
```

```python
# subplots option 2
fig, ax = plt.subplots(nrows=2, 
                       ncols=2,
                       figsize=(10, 5))
# plot to each different index
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(list(np.random.random(10)), list(np.random.random(10)));
ax[1, 0].bar(nut_butter_sales.keys(), nut_butter_sales.values());
ax[1, 1].hist(np.random.randn(1000));
```

## Which one should you use? (pyplot vs matplotlib OO method ?)

* When plotting somethig quickly, it is okay to use pyplot
* When plotting something advance, use matplotlib OO method

```python
heart_disease = df
heart_disease
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
# pyplot method
over_50.plot(kind="scatter", 
             x="age",
             y="chol",
             c="target", 
             figsize=(10, 6));
```

```python
# access all column values by data_frame.column_name.values
over_50.target.values
```

```python
# OO method mixed with 
fix, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind="scatter", 
             x="age", 
             y="chol", 
             c="target",
             ax=ax);

ax.set_xlim([45, 100]);
```

```python
# OO method from scratch
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(x=heart_disease["age"], 
                     y=heart_disease["chol"],
                     c=heart_disease["target"]
                    )

# customize the plot
ax.set(title="Colestrol Levels and Heart Disease",
       xlabel="Age",
       ylabel="Cholestrol"
      )

# Add legend
ax.legend(*scatter.legend_elements(), title="Target");

# add horizonaral dashed line
ax.axhline(heart_disease["chol"].mean(),
           ls='--'
          );
```

```python
# subplots chol, age and thalach
heart_disease.head()

fig, (ax0, ax1) = plt.subplots(nrows=2, 
                               ncols=1,
                               figsize=(10, 10),
                               sharex=True
                              )

# add some data
scatter = ax0.scatter(y=over_50["chol"], 
                      x=over_50["age"],
                      c=over_50["target"]
                     )

# customize ax0
ax0.set(title="Heart Disease and Cholestrol Level",
        xlabel="Age",
        ylabel="Chol"
       )

# add a legend
ax0.legend(*scatter.legend_elements(), title="Target")

# add a meanline
ax0.axhline(over_50["chol"].mean(),
            ls="--"
           )

# add data to ax1 plot
scatter = ax1.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"]
                     )

# customize ax1
ax1.set(title="Heart Disease and Thalach level",
#         commenting xlabel because we are sharing x axis see in main fig declaration
#         xlabel="Age",
        ylabel="Thalach Level"
       )

# add legend element
ax1.legend(*scatter.legend_elements(), title="Target")

# add horizontal line
ax1.axhline(over_50["thalach"].mean(),
            ls="--"
           );

# add super title to figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
